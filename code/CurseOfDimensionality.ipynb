```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
plt.figure(figsize=(10,10))

n_samples = 1000
dimensions = 100

points = np.random.normal(size=(n_samples, dimensions))
dists = np.linalg.norm(points.T[:,np.newaxis,:] - points.T[:,:,np.newaxis], axis=0)[np.triu_indices(n_samples,1)]
plt.hist(dists, bins=200, label="100 dimensions")

n_samples = 1000
dimensions = 200

points = np.random.normal(size=(n_samples, dimensions))
dists = np.linalg.norm(points.T[:,np.newaxis,:] - points.T[:,:,np.newaxis], axis=0)[np.triu_indices(n_samples,1)]
plt.hist(dists, bins=200, label="200 dimensions")

n_samples = 1000
dimensions = 500

points = np.random.normal(size=(n_samples, dimensions))
dists = np.linalg.norm(points.T[:,np.newaxis,:] - points.T[:,:,np.newaxis], axis=0)[np.triu_indices(n_samples,1)]
plt.hist(dists, bins=200, label="500 dimensions")

plt.title("The Curse of Dimensionality")
plt.xlabel("Distance")
plt.ylabel("Counts")
plt.legend()
plt.savefig("plots/curse_of_dim.png")
plt.show()
```

```python
for dimensions in tqdm(range(10, 501, 10)):
    plt.figure(figsize=(10,10))
    plt.xlim([0,40])
    plt.ylim([0,11000])
    n_samples = 1000
    
    points = np.random.normal(size=(n_samples, dimensions))
    dists = np.linalg.norm(points.T[:,np.newaxis,:] - points.T[:,:,np.newaxis], axis=0)[np.triu_indices(n_samples,1)]
    plt.hist(dists, bins=200, label=f"{dimensions} dimensions")
    plt.title(f"The Curse of Dimensionality")
    plt.xlabel("Distance")
    plt.ylabel("Counts")
    plt.legend()
    plt.savefig(f"curse_of_dim_gif_plots/{str(dimensions).zfill(3)}_dimensions.png")
    plt.close()
```
